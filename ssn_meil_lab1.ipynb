```python

def convert_temp(x):
  if type(x) == int or type(x) == float:
    x = (x * (9.0 / 5.0)) + 32
  else:
    print("Value provided is not a number")
  return x

def usun_duplikaty(sekwencja):
  counter = dict()
  for item in sekwencja:
    counter[item] = counter.get(item,0)+1
  for item in sorted(counter):
    print(f"{item}:{counter[item]}")

print(convert_temp(90))

sekwencja = [1,2,2,2,2,3,4,5,65,6,6,6,6]
usun_duplikaty(sekwencja)

znajdz_pierwszy = lambda sekwencja, pierwszy : sekwencja.startswith(pierwszy)

print(znajdz_pierwszy("kot", "k"))
print(znajdz_pierwszy("kot", "o"))
```

```python
class Calculator:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def set(self, x, y):
        self.x = x
        self.y = y

    def get(self):
        return self.x, self.y

    def add(self):
        return self.x + self.y

    def multiply(self):
        return self.x * self.y

    def divide(self):
        if self.y == 0:
            print("Błąd: dzielenie przez zero")
            return None
        return self.x / self.y
```

```python
class MorseTranslator:
    MORSE_MAP = {
        "a": ".-",   "b": "-...", "c": "-.-.", "d": "-..",  "e": ".",
        "f": "..-.", "g": "--.",  "h": "....", "i": "..",   "j": ".---",
        "k": "-.-",  "l": ".-..", "m": "--",   "n": "-.",   "o": "---",
        "p": ".--.", "q": "--.-", "r": ".-.",  "s": "...",  "t": "-",
        "u": "..-",  "v": "...-", "w": ".--",  "x": "-..-", "y": "-.--",
        "z": "--..",
    }

    def __init__(self, sequence: str = ""):
        self.sequence = sequence

    def set(self, sequence: str) -> None:
        self.sequence = sequence

    def get(self) -> str:
        return self.sequence

    def only_small_letters(self) -> None:
        filtered = []
        for ch in str(self.sequence):
            if ch.isalpha():
                filtered.append(ch.lower())
        self.sequence = "".join(filtered)

    def translate(self) -> str:
        self.only_small_letters()
        encoded_parts = []
        for ch in self.sequence:
            encoded_parts.append(self.MORSE_MAP.get(ch, "?"))
        morse = " ".join(encoded_parts)
        print(morse)
        return morse

```

```python
def liczby_pierwsze(upper_limit):
    primes = []
    if upper_limit < 2:
        print()
        return primes
    for candidate in range(2, upper_limit + 1):
        is_prime = True
        for divisor in range(2, candidate):
            if candidate % divisor == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(candidate)
    return primes


print(liczby_pierwsze(20))


```
