```python
import numpy as np
import pathlib
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def set_global_seed(seed_value=123):
    np.random.seed(seed_value)
    tf.random.set_seed(seed_value)

def load_flower_datasets(batch_size, image_height, image_width):
    dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
    data_path = tf.keras.utils.get_file(
        fname="flower_photos.tgz",
        origin=dataset_url,
        untar=True,
        cache_dir=".",
        cache_subdir="datasets"
    )
    data_directory = pathlib.Path(data_path).with_suffix("") / "flower_photos"
    training_dataset = tf.keras.utils.image_dataset_from_directory(
        data_directory,
        validation_split=0.2,
        subset="training",
        seed=123,
        image_size=(image_height, image_width),
        batch_size=batch_size
    )
    validation_dataset = tf.keras.utils.image_dataset_from_directory(
        data_directory,
        validation_split=0.2,
        subset="validation",
        seed=123,
        image_size=(image_height, image_width),
        batch_size=batch_size
    )
    class_names = training_dataset.class_names
    autotune_value = tf.data.AUTOTUNE
    training_dataset = training_dataset.cache().prefetch(autotune_value)
    validation_dataset = validation_dataset.cache().prefetch(autotune_value)
    return training_dataset, validation_dataset, class_names

def build_base_cnn_model(image_height, image_width, class_count):
    model = keras.Sequential([
        layers.Rescaling(1.0 / 255.0, input_shape=(image_height, image_width, 3)),
        layers.Conv2D(16, 3, activation="relu"),
        layers.MaxPooling2D(),
        layers.Conv2D(32, 3, activation="relu"),
        layers.MaxPooling2D(),
        layers.Conv2D(64, 3, activation="relu"),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(class_count)
    ])
    return model

def build_regularized_cnn_model(image_height, image_width, class_count):
    data_augmentation_layer = keras.Sequential([
        layers.RandomFlip("vertical"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
        layers.RandomContrast(0.01)
    ])
    inputs = keras.Input(shape=(image_height, image_width, 3))
    network_output = data_augmentation_layer(inputs)
    network_output = layers.Rescaling(1.0 / 255.0)(network_output)
    network_output = layers.Conv2D(16, 3, activation="relu")(network_output)
    network_output = layers.MaxPooling2D()(network_output)
    network_output = layers.BatchNormalization()(network_output)
    network_output = layers.Conv2D(32, 3, activation="relu")(network_output)
    network_output = layers.MaxPooling2D()(network_output)
    network_output = layers.BatchNormalization()(network_output)
    network_output = layers.Conv2D(64, 3, activation="relu")(network_output)
    network_output = layers.MaxPooling2D()(network_output)
    network_output = layers.BatchNormalization()(network_output)
    network_output = layers.Flatten()(network_output)
    network_output = layers.Dense(128, activation="relu")(network_output)
    network_output = layers.Dropout(0.5)(network_output)
    network_output = layers.BatchNormalization()(network_output)
    outputs = layers.Dense(class_count)(network_output)
    model = keras.Model(inputs, outputs)
    return model

def train_model(model, training_dataset, validation_dataset, optimizer_object, epoch_count, callbacks_list=None):
    loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    model.compile(optimizer=optimizer_object, loss=loss_object, metrics=["accuracy"])
    history = model.fit(
        training_dataset,
        validation_data=validation_dataset,
        epochs=epoch_count,
        callbacks=callbacks_list
    )
    return history

def plot_training_history_loss(history, title_text):
    training_loss = history.history["loss"]
    validation_loss = history.history["val_loss"]
    epoch_indices = range(1, len(training_loss) + 1)
    plt.figure()
    plt.plot(epoch_indices, training_loss, label="Training loss")
    plt.plot(epoch_indices, validation_loss, label="Validation loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(title_text)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix_for_model(model, validation_dataset, class_names, title_text):
    true_labels_list = []
    for _, label_batch in validation_dataset:
        true_labels_list.append(label_batch.numpy())
    true_labels = np.concatenate(true_labels_list, axis=0)
    logits_predictions = model.predict(validation_dataset, verbose=0)
    predicted_labels = np.argmax(logits_predictions, axis=1)
    confusion_matrix_values = confusion_matrix(true_labels, predicted_labels)
    display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_values, display_labels=class_names)
    figure, axis_object = plt.subplots(figsize=(6, 6))
    display.plot(ax=axis_object, xticks_rotation=45)
    plt.title(title_text)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    set_global_seed(123)
    batch_size = 32
    image_height = 96
    image_width = 96
    training_dataset, validation_dataset, class_names = load_flower_datasets(batch_size, image_height, image_width)
    class_count = len(class_names)

    base_model_sgd = build_base_cnn_model(image_height, image_width, class_count)
    history_sgd = train_model(
        base_model_sgd,
        training_dataset,
        validation_dataset,
        tf.keras.optimizers.SGD(),
        epoch_count=15
    )
    plot_training_history_loss(history_sgd, "Training and validation loss (SGD)")
    plot_confusion_matrix_for_model(base_model_sgd, validation_dataset, class_names, "Confusion matrix (SGD)")

    base_model_adam = build_base_cnn_model(image_height, image_width, class_count)
    history_adam = train_model(
        base_model_adam,
        training_dataset,
        validation_dataset,
        tf.keras.optimizers.Adam(),
        epoch_count=15
    )
    plot_training_history_loss(history_adam, "Training and validation loss (Adam)")
    plot_confusion_matrix_for_model(base_model_adam, validation_dataset, class_names, "Confusion matrix (Adam)")

    chosen_optimizer = tf.keras.optimizers.Adam()
    regularized_model = build_regularized_cnn_model(image_height, image_width, class_count)
    early_stopping_callback = keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True
    )
    history_regularized = train_model(
        regularized_model,
        training_dataset,
        validation_dataset,
        chosen_optimizer,
        epoch_count=30,
        callbacks_list=[early_stopping_callback]
    )
    plot_training_history_loss(history_regularized, "Training and validation loss (regularized model)")
    plot_confusion_matrix_for_model(regularized_model, validation_dataset, class_names, "Confusion matrix (regularized model)")
```
